**Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image
dengan tipe data int16.**

```python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
```

**Kemudian menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2.**

```python
# Perhatikan koordinat titik2 pt1 dan pt2
cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)

# pt1=(384,0): Ini adalah koordinat dari sudut kiri atas persegi panjang.
# (384, 0) berarti sudut kiri atas berada pada posisi x=384 dan y=0 dalam gambar.
# pt2=(510,150): Ini adalah koordinat dari sudut kanan bawah persegi panjang.
# (510, 150) berarti sudut kanan bawah berada pada posisi x=510 dan y=150 dalam gambar.
# color=(0,255,0): Ini adalah warna persegi panjang dalam format BGR (Blue, Green, Red).
# (0, 255, 0) berarti warna hijau (Green) penuh, karena komponen hijau adalah 255 sedangkan komponen biru dan merah adalah 0.
# thickness=10: Ini adalah ketebalan garis tepi persegi panjang dalam piksel.
# Nilai 10 berarti garis tepi persegi panjang akan setebal 10 piksel.
```

**Selanjutnya ditambah menambahkan bentuk persegi sesuai koordinat pt1 dan pt2
yang tertulis pada kode program.**

```python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_img)
```

**Tahap selanjutnya ditambah menambahkan bentuk lingkaran sesuai radius yang
tertulis pada kode program.**

```python
cv.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)

cv.circle(black_img,center=(400,400),radius=50,color=(255,0,0),thickness=-1) # thickness=-1 Jika ingin lingkaran tersebut diisi penuh
plt.imshow(black_img)
```

**Kemudian dilakukan penambahan garis sesuai koordinat pt1 dan pt2 sebagai berikut.**

```python
cv.line(black_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
```

**Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah
ditentukan.**

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)

# cv.LINE_AA adalah tipe garis anti-aliasing yang membuat garis lebih halus dan lebih bersih.
```

**Pembuatan black image kembali dilakukan dengan tipe data int32.**

```python
black_img2 = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
```

**Berikut adalah kode program untuk inisialisasi NumPy array dengan tipe data int32.**

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]], dtype=np.int32)
vertices
```

**Array tersebut kemudian di reshape sebagai berikut**

```python
pts = vertices.reshape((-1,1,2)) # Nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yang mewakili R, G, dan B
pts
```

**Penambahan polyline pada black image kedua yang telah dibuat.**

```python
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)

# [pts]: Ini adalah array yang berisi titik-titik untuk menggambar poligon.
# isClosed=True: Ini menunjukkan bahwa poligon harus ditutup, yaitu garis terakhir akan dihubungkan kembali ke titik awal poligon.
```
