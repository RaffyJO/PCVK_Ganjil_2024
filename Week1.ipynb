**Step 1**

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

Step 2

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://tse3.mm.bing.net/th?id=OIP.SYlcsRH0z11Vy897a3PdtgHaEj&pid=Api&P=0&h=180",
        "https://tse3.mm.bing.net/th?id=OIP.OP4RFvweCs3sDrt6eb3PEAHaE5&pid=Api&P=0&h=180",
        "https://tse1.mm.bing.net/th?id=OIP.diXA1o5n7vZDifnIghPT9QHaFB&pid=Api&P=0&h=180"]

# membaca dan menampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                          # read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) # resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  # convert color to RGB
  final_frame = cv.hconcat((image, image_2))      # concatenate image
  cv2_imshow(final_frame)                         # show image
  print('\n')
```

**Step 3**

Melihat ukuran file image

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("Resolusi image: tinggi x lebar = ", tinggi, " x ", lebar)
cv2_imshow(image_2)
```

**Step 4**

Mengakses pixel dengan memberi garis diagonal menyilang

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis horizontal di tengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

3. Pada uji coba langkah ke-9 terdapat potongan kode program sebagai berikut `cv.resize(image, (0,0), fx=0.5, fy=0.5)` Apa kegunaan kode program tersebut?dan apa pengaruhnya jika tidak dilakukan?
<br>
Jawab:
* `image`: Ini adalah variabel yang berisi data gambar yang akan diubah ukurannya.
* `(0,0)`: Parameter ini menentukan ukuran akhir gambar secara langsung (width, height). Namun, dalam kasus ini, parameter tersebut diatur ke (0,0), yang berarti ukuran spesifik tidak ditentukan, dan ukuran gambar akan ditentukan berdasarkan faktor skala fx dan fy.
* `fx=0.5`: Ini adalah faktor skala untuk lebar (width) gambar. Dengan fx=0.5, lebar gambar akan dikurangi menjadi setengah dari ukuran aslinya.
* `fy=0.5`: Ini adalah faktor skala untuk tinggi (height) gambar. Dengan fy=0.5, tinggi gambar akan dikurangi menjadi setengah dari ukuran aslinya.

```python
# Menggambar garis vertikal di tengah gambar
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(tinggi):
    image_3[y, int(lebar/2)] = [255, 255, 255]  # Garis vertikal putih di tengah

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)

# Menggambar garis menyilang diagonal pada image keluaran
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Menggambar garis diagonal dari kiri atas ke kanan bawah
for y in range(tinggi):
    x = int((lebar / tinggi) * y)  # Menyusun koordinat x agar garis diagonal tetap lurus
    image_3[y, x] = [255, 255, 255]  # Garis diagonal dari kiri atas ke kanan bawah

# Menggambar garis diagonal dari kanan atas ke kiri bawah
for y in range(tinggi):
    x = int((lebar / tinggi) * (tinggi - y - 1))  # Menyusun koordinat x untuk garis diagonal sebaliknya
    image_3[y, x] = [255, 255, 255]  # Garis diagonal dari kanan atas ke kiri bawah

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
# Buat garis horisontal berwarna putih dibagian tengah gambar dengan panjang tertentu
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(10, 51):
    image_3[int(tinggi/2), y] = [255, 255, 255]  # Garis vertikal putih di tengah

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
# Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat dalam gambar
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Tentukan batas kotak
x_start, x_end = 10, 50  # Batas kiri dan kanan kotak
y_start, y_end = 10, 50   # Batas atas dan bawah kotak

# Pastikan batas kotak berada dalam dimensi gambar (pengecekan)
x_start = max(0, x_start)
x_end = min(lebar, x_end)
y_start = max(0, y_start)
y_end = min(tinggi, y_end)

# Menggambar kotak putih pada gambar
for y in range(y_start, y_end):
    for x in range(x_start, x_end):
        image_3[y, x] = [255, 255, 255]  # Piksel putih


final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
